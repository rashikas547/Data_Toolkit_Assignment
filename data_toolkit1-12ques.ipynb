3. Explain the following terms.
a. The difference in np.array, np.asarray and np.asanyarray.
b. The difference between Deep copy and shallow copy.

np.array
Use thi when you want to ensure a new array is created, even if it means copying data from an existing array.

```python
arr = np.array([1, 2, 3], copy=True)
arr
```

np.asarray
Use this when you want to convert an input to a NumPy array but avoid unnecessary copying if the input is already an array.

```python
existing_array = np.array([1, 2, 3])
new_array = np.asarray(existing_array)
new_array
```

np.asanyarray
Use this when you want to ensure that subclasses of np.ndarray (e.g., custom array subclasses) are preserved.

```python
class MyArray(np.ndarray): pass

custom_array = MyArray([1, 2, 3])
new_array1 = np.asanyarray(custom_array)
new_array1
```

shallow_copy
Creates a new object, but if the original object has other objects inside it, it only copies the references to those inside objects.

```python
import copy
original_list = [1, 2, [3, 4]]
shallow_copy = copy.copy(original_list)

shallow_copy[2][0] = 99
print(original_list) 
```

Deep Copy
Creates a new object and also creates copies of everything inside the original object.

```python
import copy
original_list = [1, 2, [3, 4]]
deep_copy = copy.deepcopy(original_list)

deep_copy[2][0] = 99
print(original_list) 
```
